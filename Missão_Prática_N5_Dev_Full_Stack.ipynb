```python
# Geração de dados com nome e uma pontuação para cada pessoa
from faker import Faker
import random
fk = Faker('pt-br')
x=0
lista = [] 
for x in range(10):
  h = [fk.name(), random.randint(1,10)]
  s = ','.join(map(str,h))
  lista.append(s)
print(lista)
  
```

```python
# Gravando os dados em um arquivo de texto
with open('arq01.txt', 'w+') as f:
  f.writelines('\n'.join(lista))
    
```

```python
# Recuperando os dados do arquivo de texto
meu_arquivo = open('arq01.txt')
texto = meu_arquivo.read()
print(texto)
```

```python
# Manipulando os dados para que possa trabalhar com eles nas próximas etapas.
# Manipulando os dados para visualizar as pontuações das pessoas em um histograma
import matplotlib.pyplot as plt

nomes = []
pontuacao = []

f = open('arq01.txt','r')
for row in f:
  row = row.split(',')
  nomes.append(row[0])
  pontuacao.append(int(row[1]))

plt.rcParams.update({'font.size':20})
plt.figure(figsize=(10,6))
y = pontuacao
n, bins, _ = plt.hist(y, bins=10, density = True, alpha=0.7, color='b')
plt.title('Histograma das Pontuações')
plt.xlabel('Pontuações',fontsize=12)
plt.ylabel('Probabilidade',fontsize=12)
plt.grid()
plt.show()
```

```python
# Manipulando os dados para visualizar as pontuações em uma nuvem de palavras
from num2words import num2words
from wordcloud import WordCloud
por_extenso = []
for i in pontuacao:
  num = num2words(i, lang= 'pt-br')
  por_extenso.append(num)

nova_lista = ' '.join(por_extenso)

nuvem_palavras = WordCloud(background_color = 'black',width = 400, height = 200).generate(nova_lista)
plt.imshow(nuvem_palavras, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad = 0)
plt.show()
 
```
